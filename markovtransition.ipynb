```python
import random 

def markovtransition(x, probBgivenA, probAgivenB):
    if x == 'a':
        j = random.uniform(0,1)
        if j <= probBgivenA:
            return 'b'
        else:
            return 'a'
    if x == 'b':
        j = random.uniform(0,1)
        if j <= probAgivenB:
            return 'a'
        else:
            return 'b'
        
state = 'a'
probBgivenA=0.05
probAgivenB=0.50

iterations = 100

for i in range(iterations):
    print("my state at time", i, "is")
    print(state)
    state=markovtransition(state, probBgivenA, probAgivenB)
```
